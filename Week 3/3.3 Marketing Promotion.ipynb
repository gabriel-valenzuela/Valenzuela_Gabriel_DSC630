```python
import pandas as pd
import numpy as np
```

```python
baseball_df = pd.read_csv('dodgers.csv')
baseball_df.head()
```

```python
baseball_df.info()
```

```python
baseball_df['month'] = baseball_df['month'].str.replace('MAR', 'March')
baseball_df['month'] = baseball_df['month'].str.replace('APR', 'April')
baseball_df['month'] = baseball_df['month'].str.replace('MAY', 'May')
baseball_df['month'] = baseball_df['month'].str.replace('JUN', 'June')
baseball_df['month'] = baseball_df['month'].str.replace('JUL', 'July')
baseball_df['month'] = baseball_df['month'].str.replace('AUG', 'August')
baseball_df['month'] = baseball_df['month'].str.replace('SEP', 'September')
baseball_df['month'] = baseball_df['month'].str.replace('OCT', 'October')
baseball_df.head(25)
```

```python
baseball_df['cap'] = baseball_df['cap'].str.replace('NO', '0')
baseball_df['cap'] = baseball_df['cap'].str.replace('YES', '1')
baseball_df.astype({"cap": 'int64'})

baseball_df['shirt'] = baseball_df['shirt'].str.replace('NO', '0')
baseball_df['shirt'] = baseball_df['shirt'].str.replace('YES', '1')
baseball_df.astype({"shirt": 'int64'})

baseball_df['fireworks'] = baseball_df['fireworks'].str.replace('NO', '0')
baseball_df['fireworks'] = baseball_df['fireworks'].str.replace('YES', '1')
baseball_df.astype({"fireworks": 'int64'})

baseball_df['bobblehead'] = baseball_df['bobblehead'].str.replace('NO', '0')
baseball_df['bobblehead'] = baseball_df['bobblehead'].str.replace('YES', '1')
baseball_df.astype({"bobblehead": 'int64'})

```

```python
baseball_df.to_csv(r'dodgers_cleaned.csv', index = False)
```
